Which advertising channel has a strong relationship with sales volume, and can be used to model and predict the sales. 

### Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
df = pd.read_csv('Advertising.csv', index_col=0)
```

```python
# Check the columns and first few rows
df.head()
```

```python
# Get the 5-point statistics for data 
df.describe()
```

```python
# Describe the contents of this dataset
# Four variables. Each row describes an advertising campaign, a count (expenditure?) for each
# of three channels (TV, radio and newspaper), and sales volume.
```

```python
for col in df:
    df[col].plot.hist(normed=True)
    df[col].plot.kde()
    plt.legend()
    plt.show();
```

```python
# For all the variables, check if they hold normality assumption
```

```python
# Record your observations on normality here 
# TV: roughly normal/symmetrical
# Radio: roughly normal/symmetrical
# Newspaper: not normal, strongly positively skewed
# Sales: normal but positively skewed
```

### Conclusion so far !

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We shall move ahead with our analysis using TV and radio , and count out the newspaper due to the fact that data violates ols assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. We shall talk about this in the next section. 

### Step 4: Run a simple regression in `statsmodels` with TV as a predictor

```python
# import libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols
# build the formula 
f = 'sales~TV'
# create a fitted model in one line
model = ols(formula=f, data=df).fit()
```

### Step 5: Get regression diagnostics summary


```python
model.summary()
```

### Record your observations on "Goodness of fit"

```python
# R squared is 0.61, i.e. this model explains 61% of variation/61% of variance in the target
# variable can be explained using the spendings on TV, so this is an ok predictor but not great.
# Regression line is y = 0.0475 * x + 7.0326, i.e. 1 unit extra spend on TV = 0.0475 unit
# increase in sales.
```


Note here that the coefficients represent associations, not causations


### Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: We can use `model.predict()` functions to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
mm_tv = pd.DataFrame({'TV': [df.TV.min(), df.TV.max()]})
print(mm_tv)

# make predictions for those x values and store them
mm_tv_y = model.predict(mm_tv)
print(mm_tv_y)

# first, plot the observed data and the least squares line
plt.scatter(df.TV, df.sales)
plt.plot(mm_tv, mm_tv_y, color='red')
plt.xlabel('TV')
plt.ylabel('Sales');
```

### Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()
```

```python
# Record Your observations on residuals
# 'Y and fitted vs X' - positive relationship confirms that TV and sales are positiely correlated
# 'Residuals vs TV' - the variation in the residuals increases as the value of TV increases,
# i.e. the residuals are not normally distributed, indicating heteroskedasticity.
```

### Next, repeat above for radio and go through the same process, recording your observations

```python
# Model summary
f = 'sales~radio'
model = ols(formula=f, data=df).fit()
model.summary()
```

```python
print ('R-Squared:',model.rsquared)
print (model.params)
```

```python
# Plotting the data and least squares line
mm_radio = pd.DataFrame({'radio': [df.radio.min(), df.radio.max()]})
print(mm_radio)
mm_radio_y = model.predict(mm_radio)
print(mm_radio_y)
plt.scatter(df.radio, df.sales)
plt.plot(mm_radio, mm_radio_y, color='red')
plt.xlabel('Radio')
plt.ylabel('Sales');
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "radio", fig=fig)
plt.show()
```

```python
# Record your observations here for goodnes of fit 
# The model only explains 33.2% of variation in the target variable - lower than TV.
# Higher skewness and kurtosis than TV.
# A unit increase in spending on radio results in a 0.2025 unit increase in sales, which is higher
# than TV.
# Radio is also heteroskedastic.
```
